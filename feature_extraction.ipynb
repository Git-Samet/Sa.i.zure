```python
# load data
df_pp = data_import.load_pyarrow(file_name="preprocessed_df")
df_pp.columns
```

## Feature Extraction

```python
df_features = df_pp.groupby(['seizure_id', 'segment_id']).agg(['mean', 'std'])
```

## Flatten Dataframe

```python
df_features.columns = ['_'.join(col).strip() for col in df_features.columns.values]
df_features.reset_index(inplace=True)
df_features.head(20)
```
